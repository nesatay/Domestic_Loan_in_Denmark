4. Data Cleaning and Data Transformation

    In this section, we will identify, correct or remove errors and inconsistencies in data to improve data quality and then we will convert data from its original format into a format suitable for analysis.

```python
# drop unnecessary columns

columns_to_drop = ['Domestics loans from banks by data type',' currency_code',  'instrument_code', 'domestic sector code']
df_dropped = df_copy.drop(columns_to_drop, axis=1, inplace=False)
df_dropped.head(2)
```

```python
# renamed the names of some columns to make them more understandable 
 
df_renamed = df_dropped.rename(columns={'maturity (original maturity)': 
                                        'Loan Maturity', 
                                        ' currency': 'Currency' , 
                                        'instrument': 'Type of Loans', 
                                        'domestic sector': 'Sector'}, inplace=False)

df_renamed.head(2)
```

```python
melted_df = df_renamed.melt(id_vars=['Loan Maturity', 'Currency', 'Type of Loans', 'Sector'], var_name='Year_Month')
melted_df.head(2)
```

```python
# Replace '..' with '0' in the 'value' column

melted_df['value'] = melted_df['value'].replace('..', '0')
melted_df
```

```python
# Check the current data types of columns

print(melted_df.dtypes)
```

```python
# Convert 'value' column to float

melted_df['value'] = melted_df['value'].astype(float)
```

```python
#Extract year and month

melted_df['Year'] = melted_df['Year_Month'].str[:4].astype(int)
melted_df['Month'] = melted_df['Year_Month'].str[-2:].astype(int)

melted_df.head(2)
```

```python
#changed dataset name as df in order to work more easily

df = melted_df
```

```python
df['Loan Maturity'] = df['Loan Maturity'].str.replace('- ', '')

credit_period = df.groupby('Loan Maturity')['value'].sum()
credit_period
```

```python
df['Sector'] = df['Sector'].str[8:]
```

```python
df.to_csv('redi_project.csv', index=False)
```
